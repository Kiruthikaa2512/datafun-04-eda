#### Problem: You have a CBOM for a product, and multiple suppliers contribute different quantities of components. You want to visualize how much each supplier contributes to the final build.
#### Solution: Pie Chart of Supplier Component Contribution


##### Section 1. Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Sample Data: Number of components each supplier contributes to a CBOM

##### Section 2. Sample Data Set

```python
suppliers = ['Supplier A', 'Supplier B', 'Supplier C', 'Supplier D']
components = [40, 25, 20, 15]  # component counts
```

##### Section 3. Data Visualization


```python
# Plot pie chart
plt.figure(figsize=(5,5))
plt.pie(components, labels=suppliers, autopct='%1.1f%%', startangle=90)
plt.title('Supplier Contribution to CBOM Components')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.show()
```
